```python
data = {"Month":["january", "February", "March", "April", "May"]}
a = pd.DataFrame(data) #creat dictionary
print(a)
```

```python
def extract (value):  # extract first three word from colum
    return value[0:3]

a["short month"] = a["Month"].map(extract)
print(a)
```

```python
a["short month"] = a["Month"].str[0:3] #shortcut method
print(a)
```

Group By in Pandas

```python
import pandas as pd  # pip install openpyxl use this to run excel file.
data = pd.read_excel("E:/Git Demo/Python/ESD.xlsx")
print(data)
```

```python
gp = data.groupby(["Department","Gender"]).agg({"EEID":"count"})
print(gp) #used to find groupby details in the table.
```

```python
gp1 = data.groupby("Country").agg({"Age":"mean"})
print(gp1)
```

```python
gp2 = data.groupby("Country").agg({"Annual Salary":"max"})
print(gp2) #find maximum salary
```

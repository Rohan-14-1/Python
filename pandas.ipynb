Columns Transformation in pandas. Part-1

```python
df.loc[(df["Bonus"] == "0%"),"getbonus"] = "nobonus" 
df.loc[(df["Bonus"] > "0%"), "getbonus"] = "Getbonus"
print(df) # use to find the specific contant for specific colum.
```

```python
df["Add Bonus"] = df["salary"].astype(str) +" "+ df["Bonus"].astype(str)
print(df)  #using .astype(str) to add int value if we have str value then
#           no need to use.
#use ".str.Capitalize()" to print first letter capital
```

```python
data = {"Month":["january", "February", "March", "April", "May"]}
a = pd.DataFrame(data) #creat dictionary
print(a)
```

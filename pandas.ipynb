Working with missing Data in pandas

```python
print(df.dropna()) #drop null value.
```

```python
import numpy as np
print(df)
```

```python
df["salary"] = df["salary"].replace(np.nan, 30000) #Replace the missing value 
print(df)
```

```python
df["gender"] = df["gender"].replace(np.nan, "M")
print(df)
```

```python
df.drop(columns=["grnder"], inplace=True) #Drope the colume.
print(df)
```

```python
print(df["salary"].mean()) #used to find mean
```
